```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings 
warnings.filterwarnings("ignore")
import seaborn as sns
```

```python
movie = pd.read_csv("C:\\Users\\ASUS\\Documents\\Python Act\\CC104\\Movie-Ratings.csv")
```

```python
movie.columns = ['Film', 'Genre', 'RTRatings', 'ARatings','BudgetM', 'YearRelease']
```

```python
movie
```

```python
movie.RTRatings
```

```python
movie.Film = movie.Film.astype("category")
```

```python
movie.Genre = movie.Genre.astype("category")
```

```python
movie.describe()
```

```python
movie.plot()
plt.gcf().set_size_inches(15,5)
```

```python
v1 = sns.boxplot(data = movie, y = movie.BudgetM, x = movie.YearRelease)
plt.gcf().set_size_inches(15,5)
```

```python
sns.violinplot(x='YearRelease', y='BudgetM', data=movie, palette='pastel')
plt.title('Violin Plot with Seaborn')
plt.show()
```

# 1.  What is the Budget Distribution by Genre:

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Genre', y='BudgetM', data=movie, palette='muted')
plt.title('Budget Distribution by Genre')
plt.xlabel('Genre')
plt.ylabel('Budget (Millions)')
plt.xticks(rotation=45)
plt.show()
```

# 2. What is the Average Budget by Genre:

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Genre', y='BudgetM', data=movie, ci=None, palette='pastel')
plt.title('Average Budget by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Budget (Millions)')
plt.xticks(rotation=45)
plt.show()
```

# 3. What is the Average Audience Ratings (ARatings) over the Years:

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='YearRelease', y='ARatings', data=movie.groupby('YearRelease')['ARatings'].mean().reset_index(), marker = "*", ls = "--", ms =10)
plt.title('Average Audience Ratings Over the Years')
plt.xlabel('Year of Release')
plt.ylabel('Average Audience Ratings')
plt.show()
```

# 4. What is the Audience Ratings (ARatings) by Genre:

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Genre', y='ARatings', data=movie, palette='viridis')
plt.title('Audience Ratings by Genre')
plt.xlabel('Genre')
plt.ylabel('Audience Ratings')
plt.xticks(rotation=45)
plt.show()
```

# 5. Budget vs. Rotten Tomatoes Ratings (RTRatings):

```python
sns.jointplot(x='BudgetM', y='RTRatings', data=movie, kind='hex', color='blue')
plt.suptitle('Budget vs. Rotten Tomatoes Ratings', y=1.02)
plt.show()
```

# 6. What Distribution of Films Across Budget Ranges:

```python
plt.figure(figsize=(10, 6))
sns.histplot(movie, x='BudgetM', bins=[0, 50, 100, 150, float('inf')], hue='Genre', multiple='stack', palette='tab10')
plt.title('Distribution of Films Across Budget Ranges')
plt.xlabel('Budget (Millions)')
plt.ylabel('Frequency')
plt.show()
```

# 7. What is the Budget and Audience Ratings (ARatings) by Genre:

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Genre', y='BudgetM', data=movie, ci=None, color='skyblue', label='Budget')
sns.barplot(x='Genre', y='ARatings', data=movie, ci=None, color='orange', label='Audience Ratings', alpha=0.7)
plt.title('Budget and Audience Ratings by Genre')
plt.xlabel('Genre')
plt.ylabel('Values')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

# 8. What is the Average Budget Over the Years by Genre:

```python
sns.barplot(x='Genre', y='BudgetM', hue='YearRelease', data=movie, ci=None)
plt.title('Average Film Budget for Each Genre in Each Year')
plt.ylabel('Average BudgetM (in millions)')
plt.xticks(rotation=45)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

# 9. What is the Distribution of Budget Over the Years 

```python
plt.figure(figsize=(12, 6))
sns.histplot(x='YearRelease', y='BudgetM', data=movie, kde=True, palette='muted', element='poly')
plt.title('Distribution of Budget Over the Years with KDE')
plt.xlabel('Year of Release')
plt.ylabel('Budget (Millions)')
plt.show()
```

# 10. What is the Budget and Rotten Tomatoes Ratings Correlation()

```python
sns.heatmap(movie[['BudgetM', 'RTRatings']].corr(), annot=True, cmap='Blues', fmt='.2f', linewidths=1)
plt.title('Budget and Rotten Tomatoes Ratings Correlation')
plt.show()
```

# 11. What is the Average Budget by Year

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='YearRelease', y='BudgetM', data=movie, ci=None, hue='Genre', palette='Set1')
plt.title('Average Budget by Year with Hue for Genre')
plt.xlabel('Year of Release')
plt.ylabel('Average Budget (Millions)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.gcf().set_size_inches(10,5)
plt.show()
```

# 12. How do Audience Ratings vary with film budgets for each genre?

```python
sns.scatterplot(x='BudgetM', y='ARatings', hue='Genre', data=movie)
plt.title('Relationship between Film Budgets and Audience Ratings Across Genres')
plt.xlabel('BudgetM (in millions)')
plt.ylabel('ARatings')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

# 13. What is the correlation matrix between numerical columns (RTRatings, ARatings, BudgetM, YearRelease)?

```python
correlation_matrix = movie[['RTRatings', 'ARatings', 'BudgetM', 'YearRelease']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation')
plt.show()
```

# 14. What is the proportion of each genre

```python
movie['Genre'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Each Genre')
plt.axis('equal') 
plt.gcf().set_size_inches(10,8)
plt.show()
```
