```python
np.round((FreeThrows / FreeThrowAttempts) * 100)
```

```python
FTEfficiency = np.round((FreeThrows / FreeThrowAttempts) * 100)
```

# Who is the player have a highest free throw percentage??


```python
for every_player in range(len(Players)):
    plt.plot(FTEfficiency[every_player],
            color=plt.cm.viridis(every_player / len(Players)),
            ms=5,
            ls="--",
            marker="o",
            label=Players[every_player])
plt.xticks(list(range(0, 10)), Seasons)
plt.gcf().set_size_inches(8,6)
plt.legend(fontsize=6,loc="lower left", bbox_to_anchor = (1,0))

plt.show()
```

# Who is the player have a highest points free throw percentage??


```python
for every in range(len(Players)):
    plt.plot(
        FTPEfficiency[every],
        color=plt.cm.CMRmap(every / len(Players)),
        ms=5,
        ls="--",
        marker="o",
        label=Players[every]
    )
plt.xticks(list(range(0, 10)), Seasons)
plt.gcf().set_size_inches(8,6)
plt.legend(fontsize=6,loc="lower left", bbox_to_anchor = (1,0))

plt.show()
```
